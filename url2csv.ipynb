```python
import pandas as pd

#url = "http://samplecsvs.s3.amazonaws.com/SalesJan2009.csv"
#df = pd.read_csv(url)

url_path = 'http://www.anac.gov.br/assuntos/dados-e-estatisticas/percentuais-de-atrasos-e-cancelamentos-2/2018/'
url_files = ['vra-01_2018.csv',
             'vra-02_2018.csv',
             'vra_032018.csv',
             'vra_04.csv',
             'vra_05.csv',
             'vra_062018.csv',
             'vra_072018.csv',
             'vra_082018.csv',
             'vra_092018.csv']

df = pd.DataFrame()

for url_file in url_files: 
    print(url_file)
    url = url_path + url_file
    df_append = pd.read_csv(url, encoding='iso-8859-1', delimiter=';')
    df = df.append(df_append)

df.to_csv('voos.csv')

print(df.shape)
df.head()
```
